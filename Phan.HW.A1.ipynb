```python
from __future__ import print_function 
import os
from scipy.optimize import curve_fit, minimize 
import numpy as np 
import json 
import matplotlib.pyplot as plt
%pylab inline
```

```python
md = {}
means = {}
mymean = 100
np.random.seed(3009)
mysize = 2000/(np.array(range(1,100)+[10]))
```

```python
# Generate 100 Normal Distributions with same sample mean and different sample sizes

md['normal']={}
mu=mymean
sigma=1.0

for n in mysize:
    md['normal'][n] = np.random.normal(mu,sigma,n)
    
means['normal']={}
fig_mu_N = plt.figure(figsize=(10,10))
normal_mu_N = fig_mu_N.add_subplot(111)
for nn in md['normal'].iterkeys():
    means['normal'][nn] = md['normal'][nn].mean()
    normal_mu_N.plot(nn, means['normal'][nn], 'o')
    normal_mu_N.set_xlabel('sample size')
    normal_mu_N.set_ylabel('sample mean')
    normal_mu_N.plot([min(mysize), max(mysize)], [mu,mu], 'k')
```

```python
# Generate 100 Poisson Distributions with same sample mean and different sample sizes

md['poisson']={}
lam = mymean

for n in mysize:
    md['poisson'][n] = np.random.poisson(lam, n)

means['poisson']={}
fig_mu_N = plt.figure(figsize=(10,10))
poisson_mu_N = fig_mu_N.add_subplot(111)
for nn in md['poisson'].iterkeys():
    means['poisson'][nn] = md['poisson'][nn].mean()
    poisson_mu_N.plot(nn, means['poisson'][nn], 'o')
    poisson_mu_N.set_xlabel('sample size')
    poisson_mu_N.set_ylabel('sample mean')
    poisson_mu_N.plot([min(mysize), max(mysize)], [lam,lam], 'k')
```

```python
# Generate 100 Binomial Distributions with same sample mean and different sample sizes

md['binomial']={}
p = 0.5
n = mymean/p

for i in mysize:
    md['binomial'][i] = np.random.binomial(n,p,i)

means['binomial']={}
fig_mu_N = plt.figure(figsize=(10,10))
binomial_mu_N = fig_mu_N.add_subplot(111)
for ii in md['binomial'].iterkeys():
    means['binomial'][ii] = md['binomial'][ii].mean()
    binomial_mu_N.plot(ii, means['binomial'][ii], 'o')
    binomial_mu_N.set_xlabel('sample size')
    binomial_mu_N.set_ylabel('sample mean')
    binomial_mu_N.plot([min(mysize), max(mysize)], [mymean,mymean], 'k')
```

```python
# Generate 100 Chi-squared Distributions with same sample mean and different sample sizes

md['chisq']={}
df = mymean

for n in mysize: 
    md['chisq'][n] = np.random.chisquare(df, size=n)
    
means['chisq']={}
fig_mu_N = plt.figure(figsize=(10,10))
chisq_mu_N = fig_mu_N.add_subplot(111)
for nn in md['chisq'].iterkeys(): 
    means['chisq'][nn] = md['chisq'][nn].mean()
    chisq_mu_N.plot(nn, means['chisq'][nn], 'o')
    chisq_mu_N.set_xlabel('sample size')
    chisq_mu_N.set_ylabel('sample mean')
    chisq_mu_N.plot([min(mysize), max(mysize)], [df,df], 'k')
```

```python
# Generate 100 Laplace Distributions with same sample mean and different sample sizes

md['laplace']={}

loc = mymean
scale = 1.0

for n in mysize: 
    md['laplace'][n] = np.random.laplace(loc,scale,n)
    
means['laplace']={}
fig_mu_N = plt.figure(figsize=(10,10))
laplace_mu_N = fig_mu_N.add_subplot(111)
for nn in md['laplace'].iterkeys():
    means['laplace'][nn] = md['laplace'][nn].mean()
    laplace_mu_N.plot(nn, means['laplace'][nn], 'o')
    laplace_mu_N.set_xlabel('sample size')
    laplace_mu_N.set_ylabel('sample mean')
    laplace_mu_N.plot([min(mysize), max(mysize)], [loc,loc], 'k')
```

```python
# Generate 100 Logistic Distributions with same sample mean and different sample sizes

md['log']={}

loc = mymean
scale = 1.0

for n in mysize: 
    md['log'][n] = np.random.logistic(loc,scale,n)
    
means['log']={}
fig_mu_N = plt.figure(figsize=(10,10))
log_mu_N = fig_mu_N.add_subplot(111)
for nn in md['log'].iterkeys():
    means['log'][nn] = md['log'][nn].mean()
    log_mu_N.plot(nn, means['log'][nn], 'o')
    log_mu_N.set_xlabel('sample size')
    log_mu_N.set_ylabel('sample mean')
    log_mu_N.plot([min(mysize), max(mysize)], [loc,loc], 'k')
```

```python
# From the 6 distributions as displayed above, it appears that the sample mean is closer to the population mean when
# the sample size increases. 
```

```python
#Histogram of all sample means:

allmeans=[]
for n in means: 
    for i in means[n]:
        allmeans.append(means[n][i])
plt.figure(figsize=(10,10))
plt.hist(allmeans, bins=30, normed=1, color = 'orange')
plt.xlabel('Sample mean')
plt.ylabel('N')
```
